```python
import numpy as np
import matplotlib.pyplot as plt

x0 = 100  # Initial amount
p = 10  # Interest rate (%)
N = 20  # Number of years

index_set = np.arange(N + 1)
x = np.zeros(len(index_set))

# Compute solution using a loop
x[0] = x0
for n in range(1, N + 1):
    x[n] = x[n - 1] * (1 + p / 100.0)

# Print results
print(x)

# Plot results
plt.plot(index_set, x, 'ro-', label="Investment Growth")
plt.xlabel('Years')
plt.ylabel('Amount')
plt.title("Compound Interest Growth Over Time")
plt.legend()
plt.grid(True)
plt.show()
```
